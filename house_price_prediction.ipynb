```python
from sklearn.linear_model import LinearRegression
```

```python
Y = df['price']
```

```python
df['date'] = [1 if d == 2014 else 0 for d in df['date']]
```

```python
X = df.drop(['id','price'],axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = .1,random_state = 2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn import ensemble
gbr = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,learning_rate = 0.1, loss = 'squared_error')
```

```python
gbr.fit(x_train,y_train)
```

```python
gbr.score(x_test, y_test)
```
